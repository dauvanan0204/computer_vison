```python
# Image histogram
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Duong dan anh
my_path = 'D:/Python/computervision/image/01/1.jpg'
# Doc anh va hien thi
im = Image.open(my_path)
# display(im)
plt.imshow(im)
```

```python
# Chuyen sang anh xam
im_gray = im.convert('L')
plt.imshow(im_gray, cmap='gray')
```

```python
im_array = np.array(im_gray)
print(im_array)
```

```python
# To mt hnh v mi
plt.figure()
plt.hist(im_array.flatten(), bins=128)
plt.title('Biu  Histogram ca nh')
plt.xlabel('Gi tr ca pixel')
plt.ylabel('S lng')
plt.show()
```

```python
# Duong bien anh
plt.figure()
plt.gray()
plt.contour(im_gray, origin="image")
plt.show()
```
