```python
import matplotlib.pyplot as plt
```

```python
# Line chart
x = [1, 2, 3, 4, 5]
y = [10, 16, 20, 30, 25]

x1 = [1, 2, 3, 4, 5]
y1 = [20, 15, 8, 9, 20]

plt.plot(x, y, 'bo-', label = "Data1")
plt.plot(x1, y1, marker = 'o', color = 'r', linestyle = '-', label = "Data2")

plt.legend() # Chu thich
plt.xlabel('Truc x') 
plt.ylabel('Truc y')
plt.show()
```

```python
# Bar chart
categories = ['A', 'B', 'C', 'D', 'E']
values = [15, 10, 25, 12, 18]

plt.bar(categories, values, color = 'g', alpha = 0.6) # alpha la do mo

plt.legend()
plt.xlabel('Danh muc')
plt.ylabel('Gia tri')
plt.show()
```

```python
# Pie chart
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['red', 'green', 'blue', 'yellow']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title("Pie chart")
plt.show()
```

```python
# Scatter chart
x = [1, 2, 3, 4, 5]
y = [10, 16, 20, 30, 25]

plt.scatter(x, y, c='r', marker='o', label='Dữ liệu')  # Vẽ biểu đồ phân tán
plt.title('Biểu đồ phân tán ví dụ')  # Đặt tiêu đề cho biểu đồ
plt.xlabel('Trục x')  # Đặt nhãn cho trục x
plt.ylabel('Trục y')  # Đặt nhãn cho trục y
plt.legend()  # Hiển thị chú giải
plt.show()  # Hiển thị biểu đồ
```

```python
# Box chart
data = [15, 18, 22, 30, 35, 45, 50, 55, 65]

plt.boxplot(data)  # Vẽ biểu đồ hộp
plt.title('Biểu đồ hộp ví dụ')  # Đặt tiêu đề cho biểu đồ
plt.ylabel('Giá trị')  # Đặt nhãn cho trục y
plt.show()  # Hiển thị biểu đồ
```

```python
# Violin chart
data = [15, 18, 22, 30, 35, 45, 50, 55, 65]

plt.violinplot(data)  # Vẽ biểu đồ violin
plt.title('Biểu đồ violin ví dụ')  # Đặt tiêu đề cho biểu đồ
plt.ylabel('Giá trị')  # Đặt nhãn cho trục y
plt.show()  # Hiển thị biểu đồ
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = "Python is an amazing language for data visualization and word clouds."

wordcloud = WordCloud(width=800, height=400).generate(text)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Biểu đồ đám mây từ văn bản ví dụ')
plt.show()
```

```python

import numpy as np

categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 4, 3, 7, 6]

N = len(categories)
values += values[:1]

angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]

plt.polar(angles, values, 'o-', linewidth=2)
plt.fill(angles, values, 'b', alpha=0.1)
plt.title('Biểu đồ radar ví dụ')
plt.show()
```
