```python
import matplotlib.pyplot as plt
```

```python
# Line chart
x = [1, 2, 3, 4, 5]
y = [10, 16, 20, 30, 25]

x1 = [1, 2, 3, 4, 5]
y1 = [20, 15, 8, 9, 20]

plt.plot(x, y, 'bo-', label = "Data1")
plt.plot(x1, y1, marker = 'o', color = 'r', linestyle = '-', label = "Data2")

plt.legend() # Chu thich
plt.xlabel('Truc x') 
plt.ylabel('Truc y')
plt.show()
```

```python
# Bar chart
categories = ['A', 'B', 'C', 'D', 'E']
values = [15, 10, 25, 12, 18]

plt.bar(categories, values, color = 'g', alpha = 0.6) # alpha la do mo

plt.legend()
plt.xlabel('Danh muc')
plt.ylabel('Gia tri')
plt.show()
```

```python
# Pie chart
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
colors = ['red', 'green', 'blue', 'yellow']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title("Pie chart")
plt.show()
```

```python
# Scatter chart
x = [1, 2, 3, 4, 5]
y = [10, 16, 20, 30, 25]

plt.scatter(x, y, c='r', marker='o', label='D liu')  # V biu  phn tn
plt.title('Biu  phn tn v d')  # t tiu  cho biu 
plt.xlabel('Trc x')  # t nhn cho trc x
plt.ylabel('Trc y')  # t nhn cho trc y
plt.legend()  # Hin th ch gii
plt.show()  # Hin th biu 
```

```python
# Box chart
data = [15, 18, 22, 30, 35, 45, 50, 55, 65]

plt.boxplot(data)  # V biu  hp
plt.title('Biu  hp v d')  # t tiu  cho biu 
plt.ylabel('Gi tr')  # t nhn cho trc y
plt.show()  # Hin th biu 
```

```python
# Violin chart
data = [15, 18, 22, 30, 35, 45, 50, 55, 65]

plt.violinplot(data)  # V biu  violin
plt.title('Biu  violin v d')  # t tiu  cho biu 
plt.ylabel('Gi tr')  # t nhn cho trc y
plt.show()  # Hin th biu 
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = "Python is an amazing language for data visualization and word clouds."

wordcloud = WordCloud(width=800, height=400).generate(text)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Biu  m my t vn bn v d')
plt.show()
```

```python

import numpy as np

categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 4, 3, 7, 6]

N = len(categories)
values += values[:1]

angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]

plt.polar(angles, values, 'o-', linewidth=2)
plt.fill(angles, values, 'b', alpha=0.1)
plt.title('Biu  radar v d')
plt.show()
```
