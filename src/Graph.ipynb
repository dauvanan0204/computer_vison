```python
import matplotlib as matplotlib
import matplotlib.pyplot as plt

# Line chart
x = [10, 100, 1000, 10000]
y = [17300, 430400, 11244200, 32964652400]

x1 = [10, 100, 1000, 10000, 100000]
y1 = [13200, 165800, 6155100, 38665600, 1441703200]

x2 = [10, 100, 1000, 10000, 100000]
y2 = [7800, 166100, 4886900, 95545900, 8664113700]

x3 = [10, 100, 1000, 10000, 100000]
y3 = [9075100, 3650600, 7974700, 35662400, 148409500]

x4 = [10, 100, 1000, 10000, 100000]
y4 = [17500, 74000, 660900, 4490300, 24442500]

x5 = [10, 100, 1000, 10000, 100000]
y5 = [28900, 138500, 1271000, 4924700, 45105800]

plt.plot(x, y, 'b-', label = "BubbleSort")
plt.plot(x1, y1, color = 'r', linestyle = '-', label = "InsertionSort")
plt.plot(x2, y2, 'g-', label = "SelectionSort")
plt.plot(x3, y3, 'c-', label = "HeapSort")
plt.plot(x4, y4, 'y-', label = "QuickSort")
plt.plot(x5, y5, 'k-', label = "MergeSort")

plt.legend() # Chu thich
plt.xlabel('Number of element') 
plt.ylabel('Time to run(nns)')
plt.show()
```
